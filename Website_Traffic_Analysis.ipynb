# Key Performance Indicators (KPIs)

KPIs are business metrics used to evaluate website performance.

KPIs Calculated:
- Total Page Views
- Average Session Duration
- Average Bounce Rate
- Average Conversion Rate
- Total Returning Visitors

```python
total_views = df['Page Views'].sum()

print("Total Page Views:", total_views)
```

```python
avg_session = df['Session Duration'].mean()

print("Average Session Duration:", round(avg_session,2))
```

```python
avg_bounce = df['Bounce Rate'].mean()

print("Average Bounce Rate:", round(avg_bounce,2))
```

```python
avg_conversion = df['Conversion Rate'].mean()

print("Average Conversion Rate:", round(avg_conversion,2))
```

```python
returning_visitors = df[df['Previous Visits'] > 0].shape[0]

print("Returning Visitors:", returning_visitors)
```

# Traffic Source Performance Analysis

This analysis evaluates which traffic sources generate:
- Higher engagement
- Better conversion rates
- Longer session durations

```python
source_analysis = df.groupby('Traffic Source')[[
    'Session Duration',
    'Conversion Rate',
    'Page Views'
]].mean()

print(source_analysis.round(2))
```

# Returning Visitors Analysis

Returning visitors are important indicators of:

```python
returning = df[df['Previous Visits'] > 0]

new_users = df[df['Previous Visits'] == 0]

print("Returning Visitors:", returning.shape[0])

print("New Visitors:", new_users.shape[0])
```

# Feature Engineering

Feature engineering creates new columns from existing data to improve analysis.

New metrics help generate deeper business insights.

```python
df['Engagement Score'] = (
    df['Page Views'] *
    df['Session Duration'] *
    (1 - df['Bounce Rate'])
)
```

# Engagement Score Analysis

The engagement score measures overall user interaction using:
- Page views
- Session duration
- Bounce rate

Higher scores indicate stronger engagement.

```python
plt.figure(figsize=(8,5))

plt.hist(df['Engagement Score'], bins=20)

plt.title("Engagement Score Distribution")

plt.xlabel("Engagement Score")

plt.ylabel("Frequency")

plt.show()
```
